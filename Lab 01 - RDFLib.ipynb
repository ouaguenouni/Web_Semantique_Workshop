```python
myont_ns = "http://www.co-ode.org/ontologies/ont.owl#"

foaf_ns = "http://xmlns.com/foaf/0.1/"

rdf_url = "https://Data#"

def ajouter_patient(nom,prenom,nss):
    graphe = rdflib.Graph()
    
    uri_patient = rdf_url+nss
    patient_r = rdflib.URIRef(uri_patient)
    
    name_relation = rdflib.URIRef(foaf_ns+"givenName")
    familyname_relation = rdflib.URIRef(foaf_ns+"familyName")
    
    name_literal = rdflib.Literal(nom)
    familyname_literal = rdflib.Literal(prenom)
    
    graphe.add((patient_r,name_relation,name_literal))
    graphe.add((patient_r,familyname_relation,familyname_literal))
    
    return graphe
```

## Exercice récapitulatif
Dans cet exercice l'objectif est de construire une base de données RDF a partir d'un ensemble de données disponible en format CSV

```python
import pandas as pd
import numpy as np
df = pd.read_csv("netflix_titles.csv")
df = df.rename(columns={'listed_in':'categorie'})
df = df.head(100)
print(np.array(df).shape)
```

```python
def get_actors_names(cast):
    name = str(cast)
    names = name.split(",")
    l = []
    for i in names:
        n = i.split(" ")
        nn = []
        for i in n:
            if(i!=""):
                nn.append(i)
        l.append(nn)
    actors= []
    for i in l:
        actors.append( tuple([i[-1]," ".join(i[:-1])]))
    return actors

def add_actor(nom,prenom,id):
    graphe = rdflib.Graph()
    concept_t = rdflib.URIRef(myont_ns+"Acteur")
    graphe.add((concept_t,rdflib.RDF.term("type"),rdflib.OWL.term("Class")))
    acteur_r = rdflib.URIRef(rdf_url+"A"+id)
    givenname_r = rdflib.URIRef(foaf_ns+"givenName")
    familyname_r = rdflib.URIRef(foaf_ns+"familyName")
    graphe.add( (acteur_r,rdflib.RDF.term("type"),concept_t) )
    graphe.add((acteur_r,givenname_r,rdflib.Literal(prenom)))
    graphe.add( (acteur_r,familyname_r,rdflib.Literal(nom)) )
    return graphe
    
        
def create_concept_from_col(df,column,concept_name):
    graphe = rdflib.Graph()
    #Ajouter en tant que classe
    concept_t = rdflib.URIRef(myont_ns+concept_name)
    #TODO : ajouter le type du concept
    graphe.add((concept_t,rdflib.RDF.term("type"),rdflib.OWL.term("Class")))
    arr = np.array(df[column]).astype(str)
    s = set()
    #print(np.unique(arr))
    for i in np.unique(arr):
        L = i.replace(" ","").replace('.',"").replace('"',"").split(",")
        s.update(set(L))
    for i in s:
        graphe.add((rdflib.URIRef(rdf_url+i),rdflib.RDF.term("type"),concept_t))
    return graphe
```

```python

f = open("film_db.owl","w")
graphe = rdflib.Graph()

concepts_columns = ["country","categorie","type"]
datatype_columns = ["description","title","release_year"]



for i in concepts_columns:
    graphe = graphe + create_concept_from_col(df,i,i)
    pass



actors_set = set()
for i in df["cast"]:
    actors = get_actors_names(i)
    actors_set.update(set(actors))

actors_set= list(actors_set)
print("Nombre de films : ",len(actors_set))

cpt=0
for actor in actors_set:
    cpt = cpt + 1
    print(actor)
    graphe = graphe + add_actor(actor[0],actor[1],str(cpt))
    
graphe.serialize("film_db.owl","turtle")
```

```python
def ajouter_attribut(id_film,nom_propriete,valeur_propriete):
    graphe = rdflib.Graph()
    film_t = rdflib.URIRef(myont_ns+"Film")
    film_r = rdflib.URIRef(rdf_url+id_film)
    graphe.add((film_r,rdflib.RDF.term("type"),film_t))
    prop_r = rdflib.URIRef(myont_ns+nom_propriete)
    prop_v = rdflib.Literal(valeur_propriete)
    graphe.add((film_r,prop_r,prop_v))
    return graphe

def ajouter_films(df):
    graphe = rdflib.Graph()
    for i in datatype_columns:
        for idd,val in zip(df["show_id"],df[i]):
            print(val)
            graphe = graphe + ajouter_attribut(str(idd),i,val)
    return graphe

graphe = graphe + ajouter_films(df)
```

```python
def chercher_acteur(nom,prenom):
    requete = """
    prefix ns1: <http://xmlns.com/foaf/0.1/> 
    prefix ns2: <http://www.co-ode.org/ontologies/ont.owl#> 
    prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> 
    prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
    SELECT DISTINCT ?s
    WHERE{
    ?s rdf:type ns2:Acteur . 
    ?s ns1:givenName "%nom" . 
    ?s ns1:familyName "%prenom"
    }
    """.replace("%nom",nom).replace("%prenom",prenom)
    res = graphe.query(requete)
    for i in res:
        return i[0]
chercher_acteur("Adam","Pally")
```

```python
played_in = rdflib.URIRef(myont_ns+"played_in")
for acteur_list, ids in zip(df["cast"],df["show_id"]):
    film_r = rdflib.URIRef(rdf_url+str(ids))
    for nom,prenom in get_actors_names(acteur_list):
        acteur_r = chercher_acteur(prenom.replace('"','').replace('”',""),nom.replace('"','').replace('”',""))
        if(acteur_r != None):
            graphe.add((acteur_r,played_in,film_r))
```
